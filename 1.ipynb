```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, c='black', alpha=.5)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("This is my graph")
plt.grid()
```
